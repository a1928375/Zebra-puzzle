```python
import itertools

def imright(h1, h2):
    "House h1 is immediately right of h2 if h1-h2 == 1."
    
    return h1-h2 == 1

def nextto(h1, h2):
    "Two houses are next to each other if they differ by 1."
    
    return abs(h1-h2) == 1

#def zebra_puzzle2():                                                            # inefficient version
#    "Return a tuple (WATER, ZEBRA indicating their house numbers."
#    houses = first, _, middle, _, _ = [1, 2, 3, 4, 5]                          
#    orderings = list(itertools.permutations(houses)) # 1
#    return next((WATER, ZEBRA)
#                for (red, green, ivory, yellow, blue) in orderings  #第一個for先定好一暫時的sequence 
                                                                     # => 然後依序類推 => 每個for都先暫定好一暫時的sequence 
                                                                     # => 然後才run if statements
                                                                     #但有可能第一個for的sequence就不對
                                                                     # => 根本不會有答案
                                                                     #但是要等最底下的for先run完所有sequence
                                                                     #然後依序等所有的for run完所有sequence
                                                                     #第一個for才能換sequence
                                                                     # many pointless function calls
                                
#                for (Englishman, Spaniard, Ukranian, Japanese, Norwegian) in orderings
#                for (coffee, tea, milk, oj, WATER) in orderings
#                for (OldGold, Kools, Chesterfields, LuckyStrike, Parliaments) in orderings
#                for (dog, snails, fox, horse, ZEBRA) in orderings
#                   if imright(green, ivory)
#                if Englishman is red
#                if Norwegian is first
#                if nextto(Norwegian, blue)
#                if coffee is green
#                if Ukranian is tea
#                if milk is middle
#                if Kools is yellow
#                if LuckyStrike is oj
#                if Japanese is Parliaments
#                if Spaniard is dog
#                if OldGold is snails
#                if nextto(Chesterfields, fox)
#                if nextto(Kools, horse)
#                )

def zebra_puzzle():
    "Return a tuple (WATER, ZEBRA indicating their house numbers."
    
    houses = first, _, middle, _, _ = [1, 2, 3, 4, 5]                         #把houses用numbers來代表
    
    orderings = list(itertools.permutations(houses))                          #把houses的 "numbers" 來做permutation => 5!
                                                                              # orderings是"numbers"的permutation list
    
    return next((WATER, ZEBRA)                                                #找出(WATER, ZEBRA) 所對應的houses，也就是所對應的"numbers"
                
                for (red, green, ivory, yellow, blue) in orderings            #把顏色對應orderings (permutation list)，如圖            
                if imright(green, ivory)                                      #再依據if statement來做篩選
                
                for (Englishman, Spaniard, Ukranian, Japanese, Norwegian) in orderings    #把國籍對應orderings (permutation list)，如圖  
                if Englishman is red                                                      #再依據if statements來做篩選
                if Norwegian is first
                if nextto(Norwegian, blue)
                
                for (coffee, tea, milk, oj, WATER) in orderings               #把飲料對應orderings (permutation list)，如圖 
                if coffee is green                                            #再依據if statements來做篩選
                if Ukranian is tea
                if milk is middle
                
                for (OldGold, Kools, Chesterfields, LuckyStrike, Parliaments) in orderings #把香菸籍對應orderings (permutation list)，如圖
                if Kools is yellow                                                         #再依據if statements來做篩選
                if LuckyStrike is oj
                if Japanese is Parliaments
                
                for (dog, snails, fox, horse, ZEBRA) in orderings             #把寵物對應orderings (permutation list)，如圖 
                if Spaniard is dog                                            #再依據if statements來做篩選
                if OldGold is snails
                if nextto(Chesterfields, fox)
                if nextto(Kools, horse)
                )

print (zebra_puzzle())
```
