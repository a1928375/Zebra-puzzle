# 2. Instructor
# use next()
# find out one required target
# Ex: next (Knuth, Wilkes, Hamming, Minsky, Simon)  

```python
import itertools

def logic_puzzle():
    "Return a list of the names of the people, in the order they arrive."
    
    days = (mon, tue, wed, thu, fri) = (1, 2, 3, 4, 5)                           #把days用numbers來代表
    
    possible_days = list(itertools.permutations(days))                           #把days的 "numbers" 來做permutation => 5!
                                                                                 # possible_days是"numbers"的permutation list
    
    name = ["Knuth", "Wilkes", "Hamming", "Minsky", "Simon"]
    
    order =  next((Knuth, Wilkes, Hamming, Minsky, Simon)                        #找出(Knuth, Wilkes, Hamming, Minsky, Simon) 所對應的days，也就是所對應的"numbers"
                
                for (Wilkes, Hamming, Minsky, Knuth, Simon) in possible_days     #把人名對應possible_days (permutation list)，如圖
                if Knuth == Simon + 1 # 6                                        #再依據if statement來做篩選        
                
                for (programmer,writer,manager,designer,_) in possible_days      #把職業對應possible_days (permutation list)，如圖
                if Knuth == manager + 1 # 10                                     #再依據if statements來做篩選
                and thu != designer # 7
                and programmer != Wilkes and writer != Minsky # 2, 4
                
                for (laptop, droid, tablet, iphone, _) in possible_days          #把手機對應possible_days (permutation list)，如圖
                if set([laptop, Wilkes]) == set([mon, writer]) # 11              #再依據if statements來做篩選
                and set([programmer, droid]) == set([Wilkes, Hamming]) # 3
                and (iphone == tue or tablet == tue) # 12
                and designer != droid # 9
                and Knuth != manager and tablet != manager # 5
                and wed == laptop # 1
                and fri != tablet # 8
                )
    
    #print (order)
    
    #print (list(order))
    
    res = list(order)                                                            #得到一符合上述所有if statements的"numbers"
                                                                                 # Ex: [3, 1, 4, 5, 2]
                                                                                 # => 對應 name
                                                                                 # => 也就是對應(Knuth, Wilkes, Hamming, Minsky, Simon)
    
    #print (list(zip(name, res)))
    
    #print (sorted(zip(name, res), key = lambda x : x[1]))
                
    name, res = zip(*sorted(zip(name, res), key = lambda x : x[1]))              #讓name依照sequence ( x[1] ) 來排序
                                                                                 #也就是依照days來排序
    
    return list(name)

print (logic_puzzle())
```
